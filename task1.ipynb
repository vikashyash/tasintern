# Data Visualization:

# 1.HEAT MAP

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.isnull(),yticklabels=False,cmap="viridis")
```

```python
sns.set_theme("poster")
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),cmap="viridis")
```

# 2.barchart

```python
sns.set_theme("poster")
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))
sns.countplot(data=df,x="Region",hue="Channel",palette="rocket")
```

# 3. lineplot

```python
plt.figure(figsize=(15,7))
sns.set_theme("poster")
sns.set_style("darkgrid")
sns.lineplot(data=df,x="Milk",y="Fresh",hue="Channel",palette="coolwarm")
```

# 4.pie chart

```python
newdf=df.iloc[:10]
plt.figure(figsize=(20,7))
sns.set_theme("poster")
sns.set_style("darkgrid")
plt.pie(data=newdf,x="Region",labels='Channel')
plt.legend()
plt.show()
```

# 5.scatter plot

```python
regioncount1=df[df['Region']==1]
regioncount2=df[df["Region"]==2]
regioncount3=df[df["Region"]==3].iloc[0:30]
```

```python
plt.figure(figsize=(20,8))
sns.set_style("darkgrid")
plt.scatter(x=regioncount1["Milk"],y=regioncount1["Grocery"],color="red",marker="*",label="region-1")
plt.scatter(x=regioncount2["Milk"],y=regioncount2["Grocery"],color="blue",marker="+",label="region-2")
plt.scatter(x=regioncount3["Milk"],y=regioncount3["Grocery"],color="black",marker="o",label="region-3")
plt.xlabel("Milk")
plt.ylabel("Grocery")
plt.legend()
plt.show()
```

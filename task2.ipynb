```python
x=df.drop(['Outcome'],axis=1)
y=df.Outcome
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.80,test_size=0.20,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
modelfitt=rfc.fit(xtrain,ytrain)
```

```python
modelpred=modelfitt.predict(xtrain)
modelpred.reshape(274,2)
```

```python
models=[[6,148,72,35,0,33.6,0.627,50]]
modelfitt.predict(models)
```

```python
y_pred=modelfitt.predict(xtest)
from sklearn.metrics import accuracy_score
ac=accuracy_score(ytest,y_pred)
ac
```

```python
from sklearn.metrics import classification_report
cr=classification_report(ytest,y_pred)
print(cr)
```
